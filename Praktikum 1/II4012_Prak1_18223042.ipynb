```python

import re

# === JAWABAN untuk Peraturan Kemahasiswaan ITB ===

JAWABAN_JENIS_BEASISWA = [
    "ITB menyediakan dua jenis beasiswa utama, yaitu beasiswa prestasi dan beasiswa ekonomi. "
    "Bantuan bisa berupa biaya kuliah, tugas akhir, atau biaya hidup.",
    "Ada dua jenis beasiswa di ITB: beasiswa prestasi dan beasiswa ekonomi. "
    "Bentuknya macam-macam, bisa bantuan biaya kuliah sampai biaya hidup."
]

JAWABAN_TUJUAN_LAYANAN_KEMAHASISWAAN = (
    "Untuk mendukung proses pendidikan agar visi dan misi pendidikan di ITB dapat terwujud."
)

JAWABAN_JENIS_LAYANAN_KEMAHASISWAAN = (
    "Beasiswa, asrama, layanan kesehatan, bimbingan konseling, pelatihan karakter, "
    "pengembangan keprofesian, kegiatan kemahasiswaan, dan layanan disabilitas."
)

JAWABAN_IZIN_KEGIATAN = (
    "Unit Kerja ITB yang menangani urusan kemahasiswaan."
)

JAWABAN_SYARAT_BEASISWA_PRESTASI = (
    "Memiliki prestasi/potensi akademik baik (IPK baik) dan/atau capaian ko-kurikuler/ekstra-kurikuler."
)

JAWABAN_LAMA_ASRAMA = (
    "Selama-lamanya 1 (satu) tahun."
)

JAWABAN_ASAS_BK = (
    "Asas kerahasiaan, kesukarelaan, dan kode etik profesi bidang psikologi/kedokteran."
)

JAWABAN_TINGKATAN_ORGANISASI = (
    "Terpusat (KM ITB), Institut (UKM), Fakultas/Sekolah (HMJ tingkat fakultas), "
    "dan Program Studi (HMJ)."
)

JAWABAN_AFILIASI_POLITIK = (
    "Tidak, dilarang menginduk atau berafiliasi dengan organisasi luar ITB "
    "maupun partai politik manapun."
)

JAWABAN_SYARAT_PENGHARGAAN_GANESHA = (
    "Terdaftar sebagai mahasiswa ITB dan tidak sedang menerima sanksi dari ITB."
)

JAWABAN_KATEGORI_GANESHA = (
    "Ganesha Perkasa (olahraga), Ganesha Rasa (seni/budaya/agama), "
    "Ganesha Karsa (ilmiah/sosial), dan Ganesha Karya (inovatif)."
)

JAWABAN_BERHAK_ASRAMA = [
    "Layanan Asrama Mahasiswa utamanya ditujukan untuk Mahasiswa Tahap Persiapan Bersama (TPB). "
    "Mahasiswa diizinkan menghuni asrama paling lama 1 tahun.",
    "Asrama itu prioritasnya untuk anak TPB, dan batas maksimal tinggalnya adalah 1 tahun ya."
]

JAWABAN_TUJUAN_LAYANAN_KESEHATAN = (
    "Untuk membantu menjaga kesehatan dan menangani masalah kesehatan selama masa studi."
)

JAWABAN_TUJUAN_LAYANAN_DISABILITAS = (
    "Agar mereka dapat memaksimalkan potensi, menyelesaikan pendidikan dengan baik, "
    "dan berprestasi optimal."
)

JAWABAN_BENTUK_PENGEMBANGAN_KARAKTER = (
    "Studium Generale, seminar, lokakarya, pelatihan, mentoring, diskusi, "
    "latihan kepemimpinan, dan penelitian/pengabdian masyarakat."
)

JAWABAN_BENTUK_PENGEMBANGAN_KEPROFESIAN = (
    "Berupa pendidikan, pelatihan, perekrutan, dan kesempatan kerja magang."
)

JAWABAN_MINIMAL_ANGGOTA_UKM = (
    "Sekurang-kurangnya 100 orang dari minimal 3 Program Studi di Fakultas/Sekolah yang berbeda."
)

JAWABAN_LARANGAN_PENDANAAN = (
    "Dilarang menerima bantuan dari partai politik dan lembaga/organisasi "
    "yang berafiliasi dengan partai politik."
)

JAWABAN_MINIMAL_ANGGOTA_STATUS = (
    "Harus memiliki minimal 50 (lima puluh) mahasiswa sebagai anggota."
)

JAWABAN_SEMESTER_MAWAPRES = (
    "Harus sudah menyelesaikan program akademik sesuai kurikulum selama 6 (enam) semester."
)

# === RULES: (regex_pattern, jawaban) ===
rules = [
    # 1. Jenis beasiswa
    (re.compile(r"(?i).*\b(jenis|macam|apa\s+saja)\b.*\bbeasiswa\b.*"), JAWABAN_JENIS_BEASISWA),
    # 2. Tujuan layanan kemahasiswaan
    (re.compile(r"(?i).*\b(tujuan|maksud)\b.*\blayanan\s+kemahasiswaan\b.*"), JAWABAN_TUJUAN_LAYANAN_KEMAHASISWAAN),
    # 3. Jenis layanan kemahasiswaan
    (re.compile(r"(?i).*\b(jenis|macam|apa\s+saja)\b.*\blayanan\s+kemahasiswaan\b.*"), JAWABAN_JENIS_LAYANAN_KEMAHASISWAAN),
    # 4. Siapa pemberi izin kegiatan tingkat ITB
    (re.compile(r"(?i).*\b(siapa|pihak\s+mana)\b.*\b(izin\b.*\bkegiatan\b.*\btingkat\s+itb)\b.*"), JAWABAN_IZIN_KEGIATAN),
    # 5. Syarat beasiswa prestasi
    (re.compile(r"(?i).*\bsyarat\b.*\bbeasiswa\s+prestasi\b.*"), JAWABAN_SYARAT_BEASISWA_PRESTASI),
    # 6. Lama menghuni asrama
    (re.compile(r"(?i).*\b(berapa\s+lama|maksimal|batas\s+waktu)\b.*\b(menghuni|tinggal|di)\b.*\basrama\b.*"), JAWABAN_LAMA_ASRAMA),
    # 7. Asas bimbingan konseling
    (re.compile(r"(?i).*\basas\b.*\b(bimbingan|konseling|bk)\b.*"), JAWABAN_ASAS_BK),
    # 8. Tingkatan organisasi kemahasiswaan
    (re.compile(r"(?i).*\b(tingkat(an)?|jenis|macam)\b.*\borganisasi\s+kemahasiswaan\b.*"), JAWABAN_TINGKATAN_ORGANISASI),
    # 9. Afiliasi partai politik
    (re.compile(r"(?i).*\b(boleh(kah)?|aturan)\b.*\b(afiliasi|gabung|menginduk)\b.*\b(partai|politik)\b.*"), JAWABAN_AFILIASI_POLITIK),
    # 10. Syarat penghargaan Ganesha
    (re.compile(r"(?i).*\bsyarat\b.*\bpenghargaan\s+ganesha\b.*"), JAWABAN_SYARAT_PENGHARGAAN_GANESHA),
    # 11. Kategori penghargaan Ganesha
    (re.compile(r"(?i).*\b(kategori|jenis|macam)\b.*\bpenghargaan\s+ganesha\b.*"), JAWABAN_KATEGORI_GANESHA),
    # 12. Siapa berhak tinggal di asrama & berapa lama
    (re.compile(r"(?i).*\b(siapa|berhak)\b.*\b(tinggal|menghuni)\b.*\basrama\b.*\b(berapa\s+lama|batas\s+waktu)\b.*"), JAWABAN_BERHAK_ASRAMA),
    # 13. Tujuan layanan kesehatan
    (re.compile(r"(?i).*\b(tujuan|fungsi|maksud)\b.*\blayanan\s+kesehatan\b.*"), JAWABAN_TUJUAN_LAYANAN_KESEHATAN),
    # 14. Tujuan layanan disabilitas
    (re.compile(r"(?i).*\b(tujuan|maksud)\b.*\blayanan\b.*\b(disabilitas|difabel)\b.*"), JAWABAN_TUJUAN_LAYANAN_DISABILITAS),
    # 15. Bentuk kegiatan pengembangan karakter
    (re.compile(r"(?i).*\b(bentuk|jenis|kegiatan)\b.*\bpengembangan\s+karakter\b.*"), JAWABAN_BENTUK_PENGEMBANGAN_KARAKTER),
    # 16. Bentuk program pengembangan keprofesian
    (re.compile(r"(?i).*\b(bentuk|program|jenis)\b.*\bpengembangan\s+keprofesian\b.*"), JAWABAN_BENTUK_PENGEMBANGAN_KEPROFESIAN),
    # 17. Jumlah minimal calon anggota UKM baru
    (re.compile(r"(?i).*\b(jumlah|minimal|berapa)\b.*\banggota\b.*\b(ukm|unit\s+kegiatan\s+mahasiswa|baru)\b.*"), JAWABAN_MINIMAL_ANGGOTA_UKM),
    # 18. Larangan sumber pendanaan organisasi
    (re.compile(r"(?i).*\b(dilarang|tidak\s+boleh|larangan)\b.*\b(dana|bantuan|pembiayaan|pendanaan)\b.*\borganisasi\b.*"), JAWABAN_LARANGAN_PENDANAAN),
    # 19. Jumlah anggota minimal agar status tidak dicabut
    (re.compile(r"(?i).*\b(jumlah|minimal|berapa)\b.*\banggota\b.*\b(status|dicabut|pencabutan)\b.*"), JAWABAN_MINIMAL_ANGGOTA_STATUS),
    # 20. Semester untuk calon Mahasiswa Berprestasi
    (re.compile(r"(?i).*\b(berapa\s+lama|semester|syarat)\b.*\b(menempuh|kuliah|akademik)\b.*\bmahasiswa\s+berprestasi\b.*"), JAWABAN_SEMESTER_MAWAPRES),
]

fallback = "Maaf, saya tidak memahami pertanyaan Anda. Silakan tanya seputar Peraturan Kemahasiswaan ITB."

def find_answer(text: str) -> str:
    for pat, ans in rules:
        if pat.match(text):
            if isinstance(ans, list):
                return ans[0]
            return ans
    return fallback


if __name__ == "__main__":
    print("Chatbot Peraturan Kemahasiswaan ITB. Ketik 'exit' untuk keluar.")
    while True:
        user = input("\nYou: ").strip()
        if not user or user.lower() in ("exit", "quit", "keluar"):
            print("Bot: Sampai jumpa!")
            break
        print("Bot:", find_answer(user))
```

## Chatbot WITH LLM

Bagian ini mengimplementasikan *generative-based chatbot* murni menggunakan *Large Language Model* (LLM) tanpa menyertakan dokumen referensi eksternal (Non-RAG). Model dikonfigurasi melalui *system prompt* untuk bertindak sebagai asisten akademik ITB. Tujuan dari bagian ini adalah untuk mengevaluasi kemampuan bahasa natural dari model dasar, sekaligus mengamati batasan memori model serta potensi terjadinya fenomena halusinasi (mengarang fakta) ketika merespons pertanyaan yang sangat spesifik mengenai Peraturan Kemahasiswaan.

```python
import torch
import textwrap
import warnings
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

transformers.logging.set_verbosity_error()
warnings.filterwarnings("ignore")

# Menggunakan Model Qwen
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"

# Load Tokenizer dan Model
tok = AutoTokenizer.from_pretrained(MODEL_ID)
pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        device_map="auto"
    ),
    tokenizer=tok
)

def chat_once(user_msg: str) -> str:
    messages = [
        {
            "role": "system",
            "content": "Anda adalah asisten akademik virtual yang bertugas membantu mahasiswa. Jawablah pertanyaan seputar Peraturan Kemahasiswaan ITB dengan sopan, ringkas, dan menggunakan bahasa Indonesia."
        },
        {"role": "user", "content": user_msg},
    ]

    prompt = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    out = pipe(
        prompt,
        max_new_tokens=256,
        max_length=None,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        eos_token_id=tok.eos_token_id,
    )[0]["generated_text"]

    # Memotong bagian prompt agar yang di-return hanya jawaban dari bot
    reply = out[len(prompt):].strip()
    return reply

if __name__ == "__main__":
    print("Chatbot LLM Murni (Tanpa RAG). Ketik 'exit' untuk keluar.")
    while True:
        u = input("\nYou: ")
        if u.strip().lower() in ["exit", "quit", "keluar"]:
            print("Bot: Sampai jumpa!")
            break

        reply = chat_once(u)

        # wrap text
        wrapped_reply = textwrap.fill(reply, width=80)

        print(f"Bot:\n{wrapped_reply}")
```


# Chatbot RAG -LLM

Chatbot ini menggunakan dokumen PDF "Peraturan Kemahasiswaan ITB 2022".
Secara garis besar, alurnya sebagai berikut:

1. Parsing PDF → Ekstraksi teks halaman memakai pypdf.PdfReader. Teks dibersihkan (ligature, soft-hyphen, pemotongan "-\n", spasi berlebih). Dokumen dipotong per halaman sebagai chunk utama; jika halaman terlalu panjang, dipotong lagi per ukuran tetap (chunk_size=1200, overlap=200).

2. Booster Domain Kemahasiswaan → Konten diperkaya dengan booster pencarian untuk istilah kemahasiswaan: beasiswa, asrama, organisasi, UKM, HMJ, KM ITB, Ganesha, konseling, disabilitas, dll. Booster ini meningkatkan skor kandidat yang relevan saat retrieve.

3. Embedding & Indexing (RAG) → Potongan teks diubah menjadi vektor menggunakan SentenceTransformer (BAAI/bge-m3) dengan normalisasi, lalu diindeks ke FAISS (IndexFlatIP) untuk semantic search. Parameter utama: k=12 untuk pencarian awal (top-k).

4. Re-ranking (opsional) → Hasil retrieve awal dapat diurut ulang oleh CrossEncoder (BAAI/bge-reranker-v2-m3) agar konteks paling relevan naik ke atas (default topk_final=6).

5. LLM Generation → Jawaban disintesis oleh Qwen/Qwen2.5-1.5B-Instruct. Prompt menegaskan aturan: jawab singkat, jelas, dan HANYA dari konteks. Jika tidak ada informasi relevan, bot menjawab: "Tidak ditemukan dalam dokumen."

Fungsi kunci:

- read_pdf_text() → baca PDF, bersihkan teks.
- chunk_documents() → pecah teks per halaman / ukuran tetap.
- retrieve() → FAISS search + booster istilah kemahasiswaan (+ opsional re-ranking).
- make_messages() → susun prompt berisi konteks terpilih.
- generate_answer() → panggil LLM untuk menghasilkan jawaban.
- chat(query) → pipeline ujung-ke-ujung (retrieve → generate).

```python
# Cell 1 - Import libraries
import os, re, textwrap
from dataclasses import dataclass
from typing import List, Dict, Tuple
import numpy as np

from pypdf import PdfReader  # ekstraksi teks PDF berbasis teks

import faiss
from sentence_transformers import SentenceTransformer, CrossEncoder  # biarkan jika pakai re-ranking

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

print("Libraries diimport")
```

```python

# Cell 2 - Setup PDF reader — PDF-only tanpa OCR
import os, re
from pypdf import PdfReader

# Ganti ke path PDF kamu (Colab / local)
PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"

def verify_pdf(path: str) -> bool:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Tidak menemukan PDF di: {path}")
    try:
        reader = PdfReader(path)
        n_pages = len(reader.pages)
        if n_pages == 0:
            raise ValueError("PDF tidak memiliki halaman.")
        print(f"PDF OK — jumlah halaman: {n_pages}")
        return True
    except Exception as e:
        raise RuntimeError(f"Gagal membaca PDF: {e}")

def extract_pages(path: str):
    """Ekstrak teks per halaman (tanpa OCR). Halaman kosong di-skip."""
    reader = PdfReader(path)
    pages = []
    for i, page in enumerate(reader.pages, start=1):
        txt = page.extract_text() or ""
        # Normalisasi ringan
        txt = txt.replace("\u00ad", "")     # soft hyphen
        txt = txt.replace("-\n", "")        # sambung kata yang terpotong
        txt = re.sub(r"\s+", " ", txt).strip()
        if txt:
            pages.append({"page": i, "text": txt})
    print(f"Halaman dengan teks: {len(pages)}")
    return pages

# Verifikasi & contoh ekstraksi
verify_pdf(PDF_PATH)
pages_text = extract_pages(PDF_PATH)
if pages_text:
    print("Cuplikan halaman 1 (200 karakter):")
    print(pages_text[0]["text"][:200])
```

```python

# Cell 3 - Test ekstraksi teks PDF
import os

def test_pdf_extraction(path=PDF_PATH, sample_chars=300, sample_pages=3):
    """Uji ekstraksi teks dari PDF berbasis teks (tanpa OCR)."""
    if not os.path.isfile(path):
        print(f"File tidak ditemukan: {path}")
        return []
    try:
        pages = extract_pages(path)
    except Exception as e:
        print(f"Gagal ekstrak: {e}")
        return []

    n = len(pages)
    print(f"Total halaman dengan teks: {n}")
    if n == 0:
        print("Peringatan: Tidak ada teks terdeteksi. PDF kemungkinan berbasis gambar (butuh OCR).")
        return pages

    for i, p in enumerate(pages[:sample_pages], start=1):
        txt = p["text"]
        print(f"\n[Hal {p['page']}] {len(txt)} karakter")
        print(txt[:sample_chars])

    return pages

# Jalankan uji
pages_text = test_pdf_extraction(PDF_PATH)
```

```python

# Cell 4 - PDF processing (tanpa OCR, tanpa formatting ulang)

def clean_text(s: str) -> str:
    """Bersihkan teks dari karakter aneh & normalisasi spasi/baris."""
    s = (s or "")
    s = s.replace("\x00", " ").replace("\ufb01", "fi").replace("\ufb02", "fl")  # ligatures umum
    s = s.replace("\u00ad", "")        # soft hyphen
    s = s.replace("-\n", "")           # sambung kata terpotong di line break
    s = re.sub(r"\n{3,}", "\n\n", s)   # kompres newline berlebih
    s = re.sub(r"[ \t]+", " ", s)      # kompres spasi/tab
    return s.strip()

def read_pdf_text(path: str, min_chars: int = 50) -> Tuple[str, List[Tuple[int, str]]]:
    print(f"Membaca PDF: {os.path.basename(path)}")

    reader = PdfReader(path)
    pages: List[Tuple[int, str]] = []

    for i, p in enumerate(reader.pages, start=1):
        raw = p.extract_text() or ""
        t = clean_text(raw)
        pages.append((i, t))
        print(f"  Halaman {i}: {len(t)} karakter")

    suspect = [i for (i, t) in pages if len(t) < min_chars]
    if suspect:
        print(f"Peringatan: {len(suspect)} halaman terdeteksi minim teks {suspect}. "
              f"Kemungkinan berbasis gambar. Mode ini PDF-only, OCR dinonaktifkan.")

    full = "\n\n".join([f"[page {i}]\n{t}" for i, t in pages])
    total_chars = sum(len(t) for _, t in pages)
    print(f"Total: {len(pages)} halaman, {total_chars} karakter (tanpa OCR)")

    return full, pages

# Jalankan:
full_text, pages_list = read_pdf_text(PDF_PATH)
print(full_text[:500])
```

```python

# Cell 5 - Path file PDF
import os

# Ganti ke path PDF kamu (Colab / local)
PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"

def set_pdf_list(path: str):
    if os.path.isfile(path):
        print(f"File PDF ditemukan: {os.path.basename(path)}")
        return [path]
    else:
        print(f"File tidak ditemukan: {path}")
        print("Pastikan path file sudah benar.")
        return []

PDFS = set_pdf_list(PDF_PATH)
```

```python

# Cell 6 - Load & proses dokumen PDF Peraturan Kemahasiswaan ITB (tanpa formatting ulang)
import re, os

all_docs = []

CHUNK_SIZE = 1200   # ukuran maksimal per chunk (karakter)
CHUNK_OVERLAP = 200 # overlap antar chunk

def chunk_text(text: str, chunk_size: int = CHUNK_SIZE, overlap: int = CHUNK_OVERLAP) -> List[str]:
    """Potong teks menjadi chunk dengan overlap. Tanpa formatting ulang."""
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end].strip()
        if chunk:
            chunks.append(chunk)
        start += chunk_size - overlap
    return chunks

for path in PDFS:
    print(f"\n=== Proses: {os.path.basename(path)} ===")
    full_text, page_texts = read_pdf_text(path)  # fungsi dari Cell 4
    source_name = os.path.basename(path)

    for (pg, txt) in page_texts:
        if not txt.strip():
            continue

        # Jika teks halaman cukup pendek, simpan langsung sebagai satu chunk
        if len(txt) <= CHUNK_SIZE:
            all_docs.append({
                "text": txt,
                "section": f"Halaman {pg}",
                "source": source_name,
                "page_hint": pg
            })
        else:
            # Halaman panjang: potong menjadi beberapa chunk
            sub_chunks = chunk_text(txt, CHUNK_SIZE, CHUNK_OVERLAP)
            for j, sc in enumerate(sub_chunks):
                all_docs.append({
                    "text": sc,
                    "section": f"Halaman {pg} (bagian {j+1})",
                    "source": source_name,
                    "page_hint": pg
                })

print(f"\nTotal chunks: {len(all_docs)}")
if all_docs:
    print(f"Contoh chunk pertama ({len(all_docs[0]['text'])} karakter):")
    print(all_docs[0]["text"][:300])
```

```python
# Cell 7 - Setup embedding & FAISS index (PDF-only)
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

if not all_docs:
    raise ValueError("Tidak ada dokumen untuk di-embed. Pastikan Cell 6 menghasilkan 'all_docs' > 0.")

EMB_MODEL = "BAAI/bge-m3"
print(f"Memuat embedding model: {EMB_MODEL}")
embedder = SentenceTransformer(EMB_MODEL)

corpus = [d["text"] for d in all_docs]
print(f"Jumlah chunk untuk di-embed: {len(corpus)}")

# Encode ke vektor (cosine via normalize_embeddings=True → pakai IndexFlatIP)
X = embedder.encode(
    corpus,
    batch_size=64,
    normalize_embeddings=True,
    show_progress_bar=True
)
X = np.array(X, dtype="float32")

print("Membangun FAISS index (Inner Product / Cosine-normalized)...")
index = faiss.IndexFlatIP(X.shape[1])
index.add(X)

print(f"Index siap. Dimensi: {X.shape[1]}, Jumlah vektor: {index.ntotal}")
```

```python
# Cell 8 - Setup reranker (opsional tapi disarankan)
from sentence_transformers import CrossEncoder
import numpy as np

RERANKER = "BAAI/bge-reranker-v2-m3"
print(f"Memuat reranker: {RERANKER}")
reranker = CrossEncoder(RERANKER)  # otomatis pakai GPU jika tersedia (via torch)

def rerank_pairs(query: str, docs: List[Dict], topk: int = 5) -> List[Dict]:
    """
    Re-ranking doc hasil retrieval awal (mis. dari FAISS) dengan CrossEncoder.
    Mengembalikan hingga topk dokumen terbaik, menambahkan 'rerank_score'.
    """
    if not docs:
        return []
    k = min(topk, len(docs))
    pairs = [[query, d["text"]] for d in docs]
    scores = np.asarray(reranker.predict(pairs), dtype="float32")
    order = np.argsort(-scores)[:k]
    return [docs[i] | {"rerank_score": float(scores[i])} for i in order]
```

```python

# Cell 9 - Setup LLM
LLM = "Qwen/Qwen2.5-1.5B-Instruct"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(LLM, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
    LLM,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    low_cpu_mem_usage=True,
)
model.to(device)
model.eval()

def make_messages(query: str, ctxs: List[Dict]) -> List[Dict]:
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda adalah asisten yang menjawab pertanyaan tentang Peraturan Kemahasiswaan ITB. "
                "Jawab SINGKAT, JELAS, dan HANYA dari konteks. "
                "Jika tidak ada informasi relevan, balas: 'Tidak ditemukan dalam dokumen.' "
                "Jangan sertakan rujukan/sumber dalam jawaban."
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab singkat dan jelas tanpa menyertakan rujukan."
            ),
        },
    ]

@torch.inference_mode()
def generate_answer(messages, max_new_tokens=256):
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_len = inputs["input_ids"].shape[1]
    out = model.generate(
        **inputs,
        do_sample=False,
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )[0]
    gen = out[input_len:]
    txt = tokenizer.decode(gen, skip_special_tokens=True).strip()
    return re.sub(r"^assistant\s*:\s*", "", txt, flags=re.I) or "Tidak ditemukan dalam dokumen."
```

```python

# Cell 10 - Retrieval & Chat — Peraturan Kemahasiswaan ITB

from dataclasses import dataclass
from typing import List, Dict, Optional
import numpy as np
import re, os

@dataclass
class Retrieved:
    text: str
    score: float
    section: str
    source: str
    page_hint: Optional[int]

# --- Booster domain Kemahasiswaan ITB ---
KEMAHASISWAAN_TERMS = [
    "beasiswa", "prestasi", "ekonomi", "asrama", "TPB",
    "organisasi", "kemahasiswaan", "KM ITB", "UKM", "HMJ",
    "Ganesha Perkasa", "Ganesha Rasa", "Ganesha Karsa", "Ganesha Karya",
    "penghargaan", "konseling", "bimbingan", "disabilitas", "difabel",
    "layanan kesehatan", "pengembangan karakter", "pengembangan keprofesian",
    "partai politik", "pendanaan", "mahasiswa berprestasi",
    "Studium Generale", "seminar", "lokakarya", "pelatihan", "mentoring",
    "magang", "sanksi", "pencabutan", "anggota",
    "mahasiswa", "ITB", "fakultas", "sekolah", "program studi",
    "peraturan", "kegiatan", "layanan", "unit kerja",
    "afiliasi", "pendirian", "anggota",
]

IMPORTANT_TERMS_PATTERN = re.compile(
    r"\b(" + "|".join(re.escape(t) for t in KEMAHASISWAAN_TERMS) + r")\b",
    re.IGNORECASE
)

# Pattern sederhana untuk deteksi query yang kemungkinan di luar domain
_DOMAIN_KEYWORDS = re.compile(
    r"\b(beasiswa|asrama|organisasi|kemahasiswaan|mahasiswa|itb|fakultas|"
    r"sekolah|ukm|hmj|km\s*itb|ganesha|konseling|bimbingan|disabilitas|"
    r"difabel|layanan|prestasi|sanksi|peraturan|kegiatan|penghargaan|"
    r"pengembangan|keprofesian|karakter|pendanaan|anggota|magang|"
    r"studium|seminar|lokakarya|pelatihan|mentoring|mawapres|"
    r"berprestasi|tpb|kurikuler|ekstra|program\s*studi|"
    r"pencabutan|afiliasi|partai|politik|unit\s*kerja|"
    r"kesehatan|asrama|pendidikan)\b",
    re.IGNORECASE
)

def _is_domain_relevant(query: str) -> bool:
    """Cek apakah query mengandung kata kunci terkait kemahasiswaan/ITB."""
    return bool(_DOMAIN_KEYWORDS.search(query))

def _boost_ids_from_query(query: str) -> List[int]:
    """Cari doc yang mengandung istilah kemahasiswaan dari query."""
    boost_ids: List[int] = []
    if not query:
        return boost_ids

    query_terms = set(
        m.group(0).lower() for m in IMPORTANT_TERMS_PATTERN.finditer(query)
    )
    if not query_terms:
        return boost_ids

    for i, d in enumerate(all_docs):
        txt = (d.get("text") or "").lower()
        for term in query_terms:
            if term in txt:
                boost_ids.append(i)
                break
    return boost_ids

def make_messages_kemahasiswaan(query: str, ctxs: List[Dict]) -> List[Dict]:
    """Template pesan untuk LLM tentang Peraturan Kemahasiswaan ITB."""
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda adalah asisten yang HANYA menjawab pertanyaan seputar Peraturan Kemahasiswaan ITB "
                "berdasarkan KONTEKS yang diberikan.\n"
                "ATURAN KETAT yang WAJIB diikuti:\n"
                "1. DILARANG KERAS menggunakan pengetahuan umum atau informasi dari luar konteks.\n"
                "2. Jika pertanyaan TIDAK berkaitan dengan Peraturan Kemahasiswaan ITB "
                "(misalnya politik, masak, teknologi, sejarah umum, dll), "
                "WAJIB jawab HANYA: 'Tidak ditemukan dalam dokumen.'\n"
                "3. Jika jawaban tidak ada di dalam konteks yang diberikan, "
                "WAJIB jawab HANYA: 'Tidak ditemukan dalam dokumen.'\n"
                "4. Jawab SINGKAT, JELAS, dan tanpa menambahkan rujukan/sumber.\n"
                "5. JANGAN PERNAH mengarang atau menambahkan informasi yang tidak ada dalam konteks.\n"
                "6. Jika ragu, jawab: 'Tidak ditemukan dalam dokumen.'"
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab HANYA berdasarkan konteks di atas. "
                f"Jika tidak relevan atau tidak ada dalam konteks, jawab: 'Tidak ditemukan dalam dokumen.'"
            ),
        },
    ]

def retrieve(query: str, k: int = 20, topk_final: int = 8) -> List[Retrieved]:
    # 1) Booster dari istilah kemahasiswaan pada query
    bonus_ids = _boost_ids_from_query(query)

    # 2) Semantic top-k via FAISS
    qv = np.array(embedder.encode([query], normalize_embeddings=True), dtype="float32")
    D, I = index.search(qv, min(k, index.ntotal))
    I = I[0].tolist(); D = D[0].tolist()

    # 3) Gabungkan kandidat + skor awal
    cand, seen = [], set()
    for bid in bonus_ids + I:
        if bid in seen:
            continue
        seen.add(bid)
        d = all_docs[bid]
        base = D[I.index(bid)] if bid in I else 0.0
        sc = base
        if bid in bonus_ids:
            sc += 0.12
        if IMPORTANT_TERMS_PATTERN.search(d["text"]):
            sc += 0.05
        cand.append(d | {"score": float(sc)})

    if not cand:
        return []

    # 4) Re-rank (CrossEncoder) → ambil topk_final
    reranked = rerank_pairs(query, cand, topk=topk_final)
    out: List[Retrieved] = []
    for c in reranked:
        out.append(Retrieved(
            text=c["text"],
            score=float(c.get("rerank_score", c.get("score", 0.0))),
            section=c["section"],
            source=c["source"],
            page_hint=c["page_hint"]
        ))
    return out

def chat(query: str) -> Dict:
    # Gate 1: Jika query jelas di luar domain kemahasiswaan, tolak langsung
    if not _is_domain_relevant(query):
        return {"answer": "Tidak ditemukan dalam dokumen.", "contexts": []}

    ctxs = retrieve(query)

    # Gate 2: Jika semua skor reranker sangat rendah, tolak
    if ctxs and max(c.score for c in ctxs) < -2.0:
        return {"answer": "Tidak ditemukan dalam dokumen.", "contexts": []}

    msgs = make_messages_kemahasiswaan(query, [c.__dict__ for c in ctxs])
    ans = generate_answer(msgs)
    cites = [{"section": c.section, "page_hint": c.page_hint, "source": c.source} for c in ctxs]
    if ans.strip().lower() == (query or "").strip().lower():
        ans = "Tidak ditemukan dalam dokumen."
    return {"answer": ans, "contexts": cites}

print("="*60)
print("RAG Peraturan Kemahasiswaan ITB — siap dipakai!")
print("="*60)
```

```python

# Cell 11 - Chat interaktif

def start_chat(show_sources: bool = False):
    """Mulai chat interaktif untuk Peraturan Kemahasiswaan ITB."""
    print("CHATBOT Peraturan Kemahasiswaan ITB AKTIF!")
    print("Tanya apa saja tentang Peraturan Kemahasiswaan ITB.")
    print("Ketik 'exit' untuk keluar.")
    print("="*50)

    while True:
        try:
            q = input("\nAnda: ").strip()
        except EOFError:
            print("\nSesi dihentikan.")
            break

        if not q or q.lower() in {"exit", "quit", "q"}:
            print("Terima kasih!")
            break

        out = chat(q)
        print(f"\nBot: {out['answer']}")

        if show_sources and out.get("contexts"):
            print("\n[Sumber konteks]")
            for i, c in enumerate(out["contexts"], start=1):
                sec = c.get("section") or "—"
                pg  = c.get("page_hint")
                src = c.get("source") or ""
                pg_str = f"hal {pg}" if pg else "hal ?"
                print(f"{i}. {sec} ({pg_str}, {src})")

        print("-" * 50)

if __name__ == "__main__":
    start_chat(show_sources=False)
```
